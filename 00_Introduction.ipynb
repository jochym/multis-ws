## Workshop Overview

This workshop provides hands-on experience with **ab initio** (first-principles) computational methods for solid state physics and materials science. We will use the **ABINIT** package within the **Atomic Simulation Environment (ASE)** framework, running entirely in the cloud via **mybinder.org**.

### What you will learn

1. **Crystal structure building** - Create and visualize crystal structures using ASE
2. **Convergence testing** - Systematically converge numerical parameters (energy cutoff, k-point sampling)
3. **Ground state calculations** - Total energy, forces, stress tensor
4. **Phonon dispersion curves** - Lattice dynamics via finite displacements (ALAMODE/anphon)

### Target material: Silicon Carbide (3C-SiC)

- Zincblende structure (diamond-like)
- Two atoms per primitive cell (Si + C)
- Wide bandgap semiconductor
- Well-characterized experimental phonon data for validation

## How to Use This Workshop

### Running the notebooks

1. Each notebook is a sequence of **cells** - text (markdown) or code (Python)
2. **Execute a cell**: Click on it and press `Shift + Enter`
3. **Edit a cell**: Double-click on markdown cells, or click on code cells
4. **Restart kernel**: `Kernel` → `Restart Kernel` (clears all variables)
5. **Run all**: `Kernel` → `Restart Kernel and Run All Cells`

### Important notes

- **Files are not persistent** - Download your work if you want to keep it (right-click file in left panel → Download)
- **Execution time** - DFT calculations take minutes. Be patient after pressing `Shift + Enter`
- **Resource limits** - Binder provides ~2 CPU cores and limited RAM. We use reduced parameters for speed
- **Errors happen** - Read error messages carefully; they usually indicate what to fix

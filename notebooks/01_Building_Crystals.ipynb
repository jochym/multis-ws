# Building Crystal Structures: 3C-SiC

---

## Overview

In this notebook we will build the **3C-SiC (zincblende)** crystal structure using ASE.
3C-SiC is a wide bandgap semiconductor with diamond-like structure (two atoms per primitive cell).

We will:
1. Build the primitive cell
2. Create a conventional cubic cell
3. Visualize the structure
4. Export for ABINIT calculations

**Target lattice parameter**: a = 4.3812 Å (experimental)

```python
# Standard imports
import numpy as np
import matplotlib.pyplot as plt

# ASE imports
from ase.build import bulk
from ase.visualize import view
from ase.visualize.plot import plot_atoms
from ase.io import write

# Set up plotting
%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 6)
```

## 1. Primitive Cell (2 atoms)

The zincblende structure has two atoms per primitive cell (Si at 0,0,0 and C at 1/4,1/4,1/4 in fractional coordinates).

```python
# Experimental lattice parameter for 3C-SiC
a_exp = 4.3812  # Angstrom

# Build primitive cell using ASE's bulk function
siC_prim = bulk('SiC', crystalstructure='zincblende', a=a_exp, cubic=False)

print(f"Primitive cell: {len(siC_prim)} atoms")
print(f"Chemical formula: {siC_prim.get_chemical_formula()}")
print(f"Cell:\n{siC_prim.cell}")
print(f"Cell parameters: {siC_prim.cell.cellpar()}")
print(f"Positions (fractional):\n{siC_prim.get_scaled_positions()}")
print(f"Positions (Angstrom):\n{siC_prim.get_positions()}")
```

## 2. Conventional Cubic Cell (8 atoms)

For some calculations it's useful to have the conventional cubic cell (4 formula units = 8 atoms).

```python
# Build conventional cubic cell
siC_conv = bulk('SiC', crystalstructure='zincblende', a=a_exp, cubic=True)

print(f"Conventional cell: {len(siC_conv)} atoms")
print(f"Chemical formula: {siC_conv.get_chemical_formula()}")
print(f"Cell:\n{siC_conv.cell}")
print(f"Cell parameters: {siC_conv.cell.cellpar()}")
```

## 3. Visualization

Let's visualize both structures.

```python
# Plot primitive cell
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

plot_atoms(siC_prim, ax=ax[0], rotation='10x,20y,15z', radii=0.5)
ax[0].set_title('Primitive Cell (2 atoms)')
ax[0].set_axis_off()

plot_atoms(siC_conv, ax=ax[1], rotation='10x,20y,15z', radii=0.5)
ax[1].set_title('Conventional Cell (8 atoms)')
ax[1].set_axis_off()

plt.tight_layout()
plt.show()
```

## 4. Interactive 3D View (NGL)

In the binder environment you can use the interactive NGL viewer.

```python
# Interactive view - works in JupyterLab
v = view(siC_prim, viewer='ngl')
v
```

## 5. Export for ABINIT

We'll write the structure in ABINIT format for the next notebook.

```python
# Write primitive cell for ABINIT
write('siC_prim.abi', siC_prim, format='abinit')
print("Written: siC_prim.abi")

# Also write conventional cell
write('siC_conv.abi', siC_conv, format='abinit')
print("Written: siC_conv.abi")

# Show the ABINIT file content
with open('siC_prim.abi', 'r') as f:
    print(f.read())
```

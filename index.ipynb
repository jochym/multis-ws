## Environment

This workshop runs on **mybinder.org** with a two-layer setup:
- **Environment** (`env` branch): Ubuntu 24.04 + conda packages (abinit 10.0.3, ase 3.28, numpy 2.5, matplotlib 3.10, scipy 1.17, jupyterlab 4.5, nbgitpuller) + openMPI
- **Content** (`main` branch): Notebooks, data, pseudopotentials - pulled at runtime via nbgitpuller

Changing notebooks does **not** trigger environment rebuilds.
